```python
import matplotlib.pyplot as plt
```

```python
import open3d as o3d
```

```python
number_points = 150
pcd = np.random.rand(number_points, 3)
print(pcd.shape)
```

```python
# Create Figure:
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter3D(pcd[:, 0], pcd[:, 1], pcd[:, 2])
# label the axes
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Random Point Cloud")
# display:
plt.show()
```
